### Objective
Clean the raw dataset and prepare it for exploratory analysis and model training.

### Steps
1. Upload and load the dataset
2. Inspect structure and column names
3. Drop irrelevant columns
4. Encode the target variable
5. Check for missing values and duplicates
6. Investigate outliers
7. Analyze feature correlations with diagnosis
8. Save the clean dataset

```python
# Block 1 — Upload dataset from local machine
from google.colab import files
uploaded = files.upload()
```

```python
# Block 2 — Load data and first look
import pandas as pd

df = pd.read_csv('data.csv')

print('Shape:', df.shape)
df.head()
```

```python
# Block 3 — Check all column names
# Expected: 33 columns (30 features + id + diagnosis + Unnamed: 32)
print(df.columns.tolist())
```

```python
# Block 4 — Confirm Unnamed: 32 is empty and inspect id column
print("Non-null values in 'Unnamed: 32':", df['Unnamed: 32'].count())
print("Sample id values:", df['id'].head().tolist())
```

```python
# Block 5 — Drop useless columns
# 'Unnamed: 32': empty ghost column from trailing comma in CSV
# 'id': patient identifier with zero predictive value
df = df.drop(columns=['Unnamed: 32', 'id'])

print('New shape:', df.shape)
print('Columns remaining:', df.columns.tolist())
```

```python
# Block 6 — Examine target column distribution
# Class imbalance note: 62.7% Benign vs 37.3% Malignant
# This means accuracy alone is a misleading metric
print('Diagnosis value counts:')
print(df['diagnosis'].value_counts())

print('\nAs percentages:')
print(df['diagnosis'].value_counts(normalize=True).round(3) * 100)
```

```python
# Block 7 — Encode target variable
# M (Malignant) = 1 (positive class — what we want to detect)
# B (Benign)    = 0 (negative class)
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})

print('After encoding:')
print(df['diagnosis'].value_counts())
print('\nData type:', df['diagnosis'].dtype)
```

```python
# Block 8 — Full dataset health check
print('Missing values per column:')
print(df.isnull().sum())

print('\nData types:')
print(df.dtypes)
```

```python
# Block 9 — Check for duplicate rows
# In medical data, duplicates could mean a patient recorded twice
print('Duplicate rows:', df.duplicated().sum())
```

```python
# Block 10 — Statistical summary to check for outliers
# Note: area_worst and perimeter_worst will show high max values
# These are clinically valid — malignant tumours produce abnormally large cells
df.describe().round(2)
```

```python
# Block 11 — Feature correlations with diagnosis
# Identifies which features are strongest predictors of malignancy
correlations = df.corr()['diagnosis'].drop('diagnosis').sort_values(ascending=False)

print('Top 10 features most correlated with Malignancy:')
print(correlations.head(10).round(3))

print('\nTop 10 features least correlated with Malignancy:')
print(correlations.tail(10).round(3))
```

```python
# Block 12 — Save clean dataset
df.to_csv('breast_cancer_clean.csv', index=False)

print('Clean dataset saved successfully.')
print('Final shape:', df.shape)

# Block 13 — Download to local machine
from google.colab import files
files.download('breast_cancer_clean.csv')
```
